```python
data_df['HR_category'] = pd.cut(data_df['heartRate'], bins=[0, 60, 100, 150, float('inf')], labels=['niska', 'normalna', 'wysoka', 'bardzo wysoka'])

    #Niska (0-60 uderzeń na minutę): Częstość rytmu serca poniżej 60 uderzeń na minutę jest ogólnie uważana za niską i może wskazywać na bradykardię.
    #Normalna (60-100 uderzeń na minutę): To jest typowy zakres dla zdrowego dorosłego człowieka. Częstość rytmu serca w tym przedziale jest powszechnie akceptowana jako normalna.
    #Wysoka (100-150 uderzeń na minutę): Częstość rytmu serca powyżej 100 uderzeń na minutę może wskazywać na tachykardię. Zakres do 150 obejmuje zakres tachykardii, która jest częstością rytmu serca powyżej normy, ale niekiedy może być fizjologiczna, np. podczas wysiłku fizycznego.
    #Bardzo wysoka (powyżej 150 uderzeń na minutę): Częstość rytmu serca powyżej 150 uderzeń na minutę wskazuje na bardzo wysoką tachykardię, która zazwyczaj występuje w przypadku zaburzeń rytmu serca lub innych stanów patologicznych.
```

```python
#BPMeds: Jest to zwykle zmienna binarna (0 lub 1), która wskazuje, czy dana osoba przyjmuje leki przeciwnadciśnieniowe (Blood Pressure Medications).
#Wartość 1 może oznaczać, że pacjent przyjmuje leki przeciwnadciśnieniowe, natomiast wartość 0 oznacza, że nie przyjmuje.
#Leki przeciwnadciśnieniowe są stosowane w leczeniu nadciśnienia tętniczego, pomagając obniżyć ciśnienie krwi do pożądanych poziomów.

#diaBP: Skrót ten odnosi się do diastolicznego ciśnienia krwi (Diastolic Blood Pressure).
#Jest to jedna z dwóch podstawowych wartości, które są mierzone podczas pomiaru ciśnienia krwi.
#Diastoliczne ciśnienie krwi odnosi się do wartości ciśnienia krwi w naczyniach krwionośnych w czasie relaksacji lub rozkurczu serca, gdy komory serca się wypełniają krwią.
#Jest to druga liczba w standardowym odczycie ciśnienia krwi, np. 120/80 mm Hg, gdzie 80 oznacza diastoliczne ciśnienie krwi.
```

```python
# Zakresy dla podziału danych
bins_sys = [0, 90, 120, 140, 160, float('inf')]  # Przykładowe wartości graniczne dla sysBP
bins_dia = [0, 60, 80, 90, 100, float('inf')]     # Przykładowe wartości graniczne dla diaBP

# Etykiety dla przedziałów
labels = ['niskie', 'prawidłowe', 'wysokie', 'bardzo wysokie', 'nadciśnienie skrajne']

# Funkcja klasyfikująca ciśnienie krwi na podstawie sysBP i diaBP
def classify_blood_pressure(sysBP, diaBP):
    if sysBP <= 90 and diaBP <= 60:
        return 'niskie'
    elif sysBP <= 120 and diaBP <= 80:
        return 'prawidłowe'
    elif sysBP <= 140 and diaBP <= 90:
        return 'wysokie'
    elif sysBP <= 160 and diaBP <= 100:
        return 'bardzo wysokie'
    else:
        return 'nadciśnienie skrajne'

# Tworzenie nowej kolumny 'blood_pressure_category' na podstawie klasyfikacji sysBP i diaBP
data_df['blood_pressure_category'] = data_df.apply(lambda row: classify_blood_pressure(row['sysBP'], row['diaBP']), axis=1)
```
