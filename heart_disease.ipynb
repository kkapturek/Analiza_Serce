```python
data_df['HR_category'] = pd.cut(data_df['heartRate'], bins=[0, 60, 100, 150, float('inf')], labels=['niska', 'normalna', 'wysoka', 'bardzo wysoka'])

    #Niska (0-60 uderzeÅ„ na minutÄ™): CzÄ™stoÅ›Ä‡ rytmu serca poniÅ¼ej 60 uderzeÅ„ na minutÄ™ jest ogÃ³lnie uwaÅ¼ana za niskÄ… i moÅ¼e wskazywaÄ‡ na bradykardiÄ™.
    #Normalna (60-100 uderzeÅ„ na minutÄ™): To jest typowy zakres dla zdrowego dorosÅ‚ego czÅ‚owieka. CzÄ™stoÅ›Ä‡ rytmu serca w tym przedziale jest powszechnie akceptowana jako normalna.
    #Wysoka (100-150 uderzeÅ„ na minutÄ™): CzÄ™stoÅ›Ä‡ rytmu serca powyÅ¼ej 100 uderzeÅ„ na minutÄ™ moÅ¼e wskazywaÄ‡ na tachykardiÄ™. Zakres do 150 obejmuje zakres tachykardii, ktÃ³ra jest czÄ™stoÅ›ciÄ… rytmu serca powyÅ¼ej normy, ale niekiedy moÅ¼e byÄ‡ fizjologiczna, np. podczas wysiÅ‚ku fizycznego.
    #Bardzo wysoka (powyÅ¼ej 150 uderzeÅ„ na minutÄ™): CzÄ™stoÅ›Ä‡ rytmu serca powyÅ¼ej 150 uderzeÅ„ na minutÄ™ wskazuje na bardzo wysokÄ… tachykardiÄ™, ktÃ³ra zazwyczaj wystÄ™puje w przypadku zaburzeÅ„ rytmu serca lub innych stanÃ³w patologicznych.
```

```python
#BPMeds: Jest to zwykle zmienna binarna (0 lub 1), ktÃ³ra wskazuje, czy dana osoba przyjmuje leki przeciwnadciÅ›nieniowe (Blood Pressure Medications).
#WartoÅ›Ä‡ 1 moÅ¼e oznaczaÄ‡, Å¼e pacjent przyjmuje leki przeciwnadciÅ›nieniowe, natomiast wartoÅ›Ä‡ 0 oznacza, Å¼e nie przyjmuje.
#Leki przeciwnadciÅ›nieniowe sÄ… stosowane w leczeniu nadciÅ›nienia tÄ™tniczego, pomagajÄ…c obniÅ¼yÄ‡ ciÅ›nienie krwi do poÅ¼Ä…danych poziomÃ³w.

#diaBP: SkrÃ³t ten odnosi siÄ™ do diastolicznego ciÅ›nienia krwi (Diastolic Blood Pressure).
#Jest to jedna z dwÃ³ch podstawowych wartoÅ›ci, ktÃ³re sÄ… mierzone podczas pomiaru ciÅ›nienia krwi.
#Diastoliczne ciÅ›nienie krwi odnosi siÄ™ do wartoÅ›ci ciÅ›nienia krwi w naczyniach krwionoÅ›nych w czasie relaksacji lub rozkurczu serca, gdy komory serca siÄ™ wypeÅ‚niajÄ… krwiÄ….
#Jest to druga liczba w standardowym odczycie ciÅ›nienia krwi, np. 120/80 mm Hg, gdzie 80 oznacza diastoliczne ciÅ›nienie krwi.
```

```python
# Zakresy dla podziaÅ‚u danych
bins_sys = [0, 90, 120, 140, 160, float('inf')]  # PrzykÅ‚adowe wartoÅ›ci graniczne dla sysBP
bins_dia = [0, 60, 80, 90, 100, float('inf')]     # PrzykÅ‚adowe wartoÅ›ci graniczne dla diaBP

# Etykiety dla przedziaÅ‚Ã³w
labels = ['niskie', 'prawidÅ‚owe', 'wysokie', 'bardzo wysokie', 'nadciÅ›nienie skrajne']

# Funkcja klasyfikujÄ…ca ciÅ›nienie krwi na podstawie sysBP i diaBP
def classify_blood_pressure(sysBP, diaBP):
    if sysBP <= 90 and diaBP <= 60:
        return 'niskie'
    elif sysBP <= 120 and diaBP <= 80:
        return 'prawidÅ‚owe'
    elif sysBP <= 140 and diaBP <= 90:
        return 'wysokie'
    elif sysBP <= 160 and diaBP <= 100:
        return 'bardzo wysokie'
    else:
        return 'nadciÅ›nienie skrajne'

# Tworzenie nowej kolumny 'blood_pressure_category' na podstawie klasyfikacji sysBP i diaBP
data_df['blood_pressure_category'] = data_df.apply(lambda row: classify_blood_pressure(row['sysBP'], row['diaBP']), axis=1)
```
